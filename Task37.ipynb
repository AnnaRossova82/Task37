```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sb
sb.set_style('whitegrid')
import matplotlib.pyplot as plt
import sklearn
from sklearn.preprocessing import scale
from pylab import rcParams
rcParams['figure.figsize']=10,8
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
```

```python
data=pd.read_csv('Position_Salaries.csv')
data.head()
X1 = data['Level']
y1 = data['Salary']
X=np.array(X1).reshape(-1,1)
y=np.array(y1)
#plt.scatter(X,y,s=10)
#plt.show()
model=LinearRegression()
model.fit(X,y)
model.score(X,y)
print("Scores: ",model.score(X,y))
y_pred=model.predict(X)
polynomial_features=PolynomialFeatures(degree=7)
X_poly=polynomial_features.fit_transform(X)
```

```python
model=LinearRegression()
model.fit(X_poly,y)
model.score(X_poly,y)
print("Scores: ",model.score(X_poly,y))
y_poly_pred=model.predict(X_poly)
```

```python
plt.scatter(X,y,s=10)
plt.plot(X_poly,y_pred,color='r')
plt.show()
```

```python
#Для нормалізації можна видалити не функціональні колонки, змінити значення на середні, в даному датасеті не придумала
#що зробити окрім замінити текстові дані і знаки питання на нульові. Ще з буде лекція з нормалізації здається.


df=pd.read_csv('communities.txt')
#df.to_numpy()
print(df.head())
# Checking for null values
#print (df.isnull())
df.dtypes.value_counts()
# Droping the unnecessary Columns
#df=df.drop (df.columns[[1,2,10,14,15,16,17,19]], axis= 1, inplace= True)
#df
df['Lakewoodcity']=pd.to_numeric(df['Lakewoodcity'],errors='coerce')
df = df.replace(['?'],'0.00')
df['Lakewoodcity'] = df['Lakewoodcity'].fillna(0)
df[['?']] = df[['?']].fillna(df[['?']].mean())
df



```

```python
y=df[['8']]
y=np.array(y1).reshape(-1,1)
X1=df[df.columns.difference(['8'])]
X=np.array(X1)
#sklearn.utils.check_X_y(X, y, accept_sparse=False, *, accept_large_sparse=True, dtype='numeric', order=None, copy=False, force_all_finite=True, ensure_2d=True, allow_nd=False, multi_output=False, ensure_min_samples=1, ensure_min_features=1, y_numeric=False, estimator=None)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
lr=LinearRegression().fit(X_train,y_train)
print(lr.score(X_train,y_train))
lr=LinearRegression().fit(X_test,y_test)
print(lr.score(X_test,y_test))

```

```python
ridge=Ridge(alpha=0.9).fit(X_train,y_train)
print(ridge.score(X_train,y_train))
ridge=Ridge(alpha=0.9).fit(X_test,y_test)
print(ridge.score(X_test,y_test))
```

```python
lasso=Lasso(alpha=0.9).fit(X_train,y_train)
print(ridge.score(X_train,y_train))
lasso=Lasso(alpha=0.9).fit(X_test,y_test)
print(ridge.score(X_test,y_test))
```

```python
elast_net=ElasticNet(alpha=0.01, l1_ratio=0.01).fit(X_train,y_train)
print(elast_net.score(X_train,y_train))
elast_net=ElasticNet(alpha=0.01, l1_ratio=0.01).fit(X_test,y_test)
print(elast_net.score(X_test,y_test))
```
